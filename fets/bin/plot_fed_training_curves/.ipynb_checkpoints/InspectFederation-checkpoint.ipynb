```python
############# Testing equality of fast and slower log parsing
```

```python
#### Initial run with lr=4e-5 and epochs per round = 1/7 (7 workers, Adam)
round_nums_, validation_scores_ = parse_logs(running=False, specific_directory='Completed/agg_20201014-214253', faster=False)
```

```python
#### Initial run with lr=4e-5 and epochs per round = 1/7 (7 workers, Adam)
alt_round_nums_, alt_validation_scores_ = parse_logs(running=False, specific_directory='Completed/agg_20201014-214253', faster=True)
```

```python
round_val_dict = dict(zip(round_nums_, validation_scores_))
alt_round_val_dict = dict(zip(alt_round_nums_, alt_validation_scores_))
fail = False
for round, val in round_val_dict.items():
    if val != alt_round_val_dict[round]:
        print("--ERROR--")
        fail=True
if not fail:
    print("Arrays were equal :)")
```

```python
#### This one with same lr but epochs per round = 0.02
round_nums, validation_scores = parse_logs(running=False, specific_directory='Completed/agg_20201016-100954', faster=False)
```

```python
#### This one with same lr but epochs per round = 0.02
alt_round_nums, alt_validation_scores = parse_logs(running=False, specific_directory='Completed/agg_20201016-100954', faster=True)
```

```python
round_val_dict = dict(zip(round_nums, validation_scores))
alt_round_val_dict = dict(zip(alt_round_nums, alt_validation_scores))
fail = False
for round, val in round_val_dict.items():
    if val != alt_round_val_dict[round]:
        print("--ERROR--")
        fail=True
if not fail:
    print("Arrays were equal :)")
```
